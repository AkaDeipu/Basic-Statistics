#**Standardization of Values**

**__Explain the concept of standardization (z-score normalization).__**

Standardization is a technique used to transform numerical data or features in the data set to have mean of 0 and standard deviation of 1. This is a preprocessing technique which makes all the features to contribute equally while creating a model. Standardization makes sure that features with different ranges or units don't unfairly affect the model's results and helps reduce the effect of extreme values on a model by adjusting the data to have a balanced, centered spread.



```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(num_col), columns = num_col.columns)
scaled_df
```

# Importance of standardization and one-hot encoding.

**Standardization** It centers all numeric features around 0 with a standard deviation of 1. This Makes all number based features have the same scale, so no single feature unfairly influences the model. It also reduces the impact of extreme values and helps the model learn faster.

**One Hot Encoding** It converts categorical variables into a format computers understand, using 0s and 1s. This prevents the model from assuming false relationships between categories. By representing categories as separate binary features, the model can clearly distinguish the effect of each category on the outcome.

By these steps we can improve model accuracy, reduce training time, handles diverse data and enables fair comparision.
